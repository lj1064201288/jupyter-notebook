```python
# 汉诺塔的实现
def hano(n, a, b, c):
    '''
    汉诺塔的递归实现
    n:　代表几个盘子
    a:　代表第一个塔，开始的塔
    b:  代表第二个塔，中间的塔
    c:  代表第三个塔，目标塔
    '''
    if n == 1:
        print(a, "－>", c)
        return None
    #　把n-1个盘子，从a塔借助于c塔，挪到b塔上去
    hano(n-1, a, c, b)
    print(a, "－>", c)
    #　把n-1个盘子，从b塔，记住与a塔，挪到c塔上去
    hano(n-1, b, a, c)
    
a = 'A'
b = 'B'
c = 'C'

n = 1
hano(n, a, b, c)
```

```python
n = 2
hano(n, a, b, c)
```

```python
n = 3
hano(n, a, b, c)
```

```python
n = 10
hano(n, a, b, c)
```

# List(列表)
    -del: 命令

```python
#　del删除
a = [1,2,3,4,5,6]
del a[2]
print(a)
```

```python
#　del　删除
# 如果使用del之后,id的值和删除前不一样，则说明删除生成了一个新的list
a = [1,2,3,4,5,6]
print(id(a))
del a[2]
print(id(a))
print(a)
```

```python
# del 一个变量后不能再继续使用此变量
del a 
print(a)
```

```python
#　使用加号连接两个列表
a = [1,2,3,4,5]
b = [5,6,7,8,9,0]
d = ['a','b','c']
c = a+b+d
print(c)
```

```python
#　使用乘号操作列表
# 列表直接跟一个整数相乘
#　相当于把n个列表接在一起
a = [1,2,3,4,5]
b =a*3
print(b)
```

```python
#　成员资格运算
#　就是判断一个元素是否在list里边
a = [1,2,3,4,56,7,]
b = 8
# c 的值是一个布尔值
c = b in a
print(c)
b = 4
print(b in a)
```

```python
#　not　in
a = [1,2,3,4,5]
b = 9
print(b not in a)
```

# 列表的遍历
    -for
    -while

```python
# for in list
a = [1,2,3,4]

#　挨个打印a里面的元素
for i in a:
    print(i)
```

```python
#　jave,c++程序员写的python代码是这样的
for i in range(0,len(a)):
    print(a[i])
    i += 1
```

```python
b = ["I love wangxiaojing"]

for i in b:
    print(i)
```

```python
# range
# in 后面的变量要求是可以迭代的内容
for i in range(1,10):
    print(i)
    
print(type(range(1,10)))
```

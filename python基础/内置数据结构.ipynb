```python
# 汉诺塔的实现
def hano(n, a, b, c):
    '''
    汉诺塔的递归实现
    n:　代表几个盘子
    a:　代表第一个塔，开始的塔
    b:  代表第二个塔，中间的塔
    c:  代表第三个塔，目标塔
    '''
    if n == 1:
        print(a, "－>", c)
        return None
    #　把n-1个盘子，从a塔借助于c塔，挪到b塔上去
    hano(n-1, a, c, b)
    print(a, "－>", c)
    #　把n-1个盘子，从b塔，记住与a塔，挪到c塔上去
    hano(n-1, b, a, c)
    
a = 'A'
b = 'B'
c = 'C'

n = 1
hano(n, a, b, c)
```

```python
# range
# in 后面的变量要求是可以迭代的内容
for i in range(1,10):
    print(i)
    
print(type(range(1,10)))
```

```python
#　循环while访问list
# 一般不使用while访问list

a = [1,2,3,4,5,6,]
lenghth = len(a)
#　index表示的是list的下标
index = 0
while index < lenghth:
    print(a[index])
    index += 1
```

```python
#　双层列表循环
#　a 为嵌套列表，或者叫双层列表
a = [["one", 1], ["two", 2], ["three", 3]]

for k,v in a:
    print(k, "——————", v)
```

```python
#　双层列表循环变异
# a为嵌套列表，或者叫做双层列表
a = [["one", 1], ["two", 2], ["three", 3,4,5,6,7,8]]

for k,v in a:
    print(k, "————————————————", v)
```

```python
#　双层列表循环变异
#　a 为乔涛列表,或者叫双层列表
a = [["one", 1, "eins"], ["two", 2, "zwei"], ["three", 3, "drei"]]
#　这个例子说明，k,v,w的个数应该跟解包出来的变量个数一致
for k,v,w in a:
    print(k, "——————", v, "——————", w)
```

# 列表内涵：list content
    -　通过简单的方法创作列表

```python
#　for 创建
a = ['a', 'b', 'c']
#用list a 创建一个list　b
# 下面代码的含义是，对于所有a中的元素，逐个放入新列表中
b = [i for i in a]
print(b)
```

```python
#　对a中的所有元素乘以１０,生成一个新的list
a = [1,2,3,4,5]
#　用list a 创建一个list　b
#　下面的代码含义是，对于所有a中的元素，逐个放入新列表b中
b = [i*10 for i in a]
print(b)
```

```python
#　还可以过滤原来list中的内容并放入新列表
#　比如原有列表a,　需要把所有的a中的偶数生成新的列表b
a = [x for x in range(1,300)]#生成1到299的一个列表
b = [m for m in a if m%2 == 0]
print(b)
```

```python
#　列表生成式可以嵌套
#　由两个列表a, b
a = [i for i in range(1,4)] #生成list a
print(a)

b = [i for i in range(100,400) if i%100 == 0 ]
print(b)
#　列表生成是可以嵌套的，此时等于两个for循环嵌套
c = [m+n for m in a for n in b]
print(c)

#上面的代码跟下面代码等价
for m in a:
    for n in b:
        print(m+n,end=" ")
print()

#　嵌套的列表生成式也可以用条件表达式
c = [m+n for m in a for n in b if m+n < 250]
print(c)
```

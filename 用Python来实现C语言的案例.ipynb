```python
# 九九乘法表
for row in range(1,10):
    for i in range(1,row+1):
        print(str(i) + '*' + str(row) + '=',end=' ')
        print(i*row, end=' ')
    print(' ')
```

```python
# 斐波拉契数列

a = 1
b = 1

while a < 10000:
    print(a, end=' ')
    print(b, end=' ')
    a += b
    b += a
```

```python
# 斐波拉契递归排列
def fab(n):
    if n == 1:
        return 1
    if n == 0:
        return 0
    else:
        result = int(fab(n-1)) + int(fab(n-2))
        return result

    
for i in range(10):
    print(fab(i))
```

```python
# 冒泡排序
a = [1,4,6,3,612,36,73,56,34,23]
b = len(a)
for i in range(b):
    for j in range(b):
        if a[i] < a[j]:
            t = a[i]
            a[i] = a[j]
            a[j] = t
print(a)
```

```python
# 快速排序
a = [1,4,6,3,612,36,73,56,34,23]

def sort_quick(lists, low, hige):
    if low >= hige:
        return lists
    
    lift = low
    right = hige
    key = lists[low]
    while lift < right:
        while lift < right and a[right] >= key:
            right -= 1
        a[lift] = a[right]
        while lift < right and a[lift] <= key:
            lift += 1
        a[right] = a[lift]
        a[lift] = key
        sort_quick(lists, low, lift-1)
        sort_quick(lists, lift+1, hige)
        return lists
b = len(a)-1
sort_quick(a, 0, b)
    
```

```python
a = [1,4,6,3,612,36,73,56,34,23]

def insert_sort(lists):
    count = len(lists)
    for i in range(1, count):
        key = lists[i]
        j = i - 1
        while j >= 0:
            if lists[j] > key:
                lists[j+1] = lists[j]
                lists[j] = key
            j -= 1
    return lists

insert_sort(a)
```

```python
import threading
import time

def fun1():
    print("Starting1....",time.ctime())
    time.sleep(4)
    print("Ending1....",time.ctime())
    
def fun2():
    print("Starting2....",time.ctime())
    time.sleep(2)
    print("Ending2...",time.ctime())
    
def fun():
    print("Start....",time.ctime)
    t1 = threading.Thread(target=fun1, args=())
    t2 = threading.Thread(target=fun2, args=())
    
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()
    
    print("End....",time.ctime())
fun()
```

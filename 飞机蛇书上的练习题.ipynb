```python
# 用户
class User():
    
    def __init__(self, first, last, age, num, attempt = " "):
        self.first_name = first
        self.last_name = last
        self.old = age
        self.number = num
        self.login_attempts = 0
    
    def increment_login_attempts(self):
        self.login_attempts += 1
        print(str(self.login_attempts))
        
    def reset_login_attempts(self):
        self.login_attempts = 0
        print(str(self.login_attempts))
    
    def describe_user(self):
        print("用户的名字是:", self.first_name + self.last_name)
        print("用户" + str(self.old) + "岁了!")
        print("用户的电话是:",str(self.number))
    def greet_user(self):
        print("Hello," + self.first + self.last)

user = User("刘","俊",18, 13550211725)
print(user.first_name,user.last_name)
user.describe_user()
user.increment_login_attempts()
user.increment_login_attempts()
user.reset_login_attempts()
```
